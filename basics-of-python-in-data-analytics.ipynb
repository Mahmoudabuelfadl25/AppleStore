```python
import pandas as pd 
```

```python
df =pd.read_csv(r"/kaggle/input/orders/Orders.O.csv",skiprows=4)
df
```

```python
# To delet rows that it's all cells is null 
df.dropna(how="all")
```

```python
# to Capital the first letter in the city column 
df["City"].str.title()
```

```python
# To apply it in the main table 
df["City"] = df["City"].str.title()
df
```

```python
# if you want to upper branch letters 
df["Branch"] = df["Branch"].str.upper()
```

```python
# if you want to small Email letter 
df['Email'] = df['Email'].str.lower()
```

```python
df
```

```python
# To replace any word , such as Tel : by " " to remove it 
df["Phone Number"] = df["Phone Number"].str.replace("Tel: "," ")
df
```

```python
# (0) No dicimal number , (1) Decimal Num  
df["Discount"] = df["Discount"].round(0)
df
```

```python
# to make new column resulted from 2 column operation df['Tal Price'] = df["Quantity"] * df["Unit Price"]
df["Total Price"] = df["Unit Price"] * df['Quantity']
#to replace na value in discount column by 0 .. 
df['Cost'] = df['Total Price'] - df['Discount'].fillna(0)
df
```

```python
# if you want to insert column in sepecific place in the table,9 is the rank of column from the left
df.insert(9,"Full Name",df["Customer First Name"]+ " "+ df["Customer Last Name"])
df
```

```python
# to take th first 2 letters only from any column and put it in new one 
NewColumn = df['Branch'].str[:2]
# Now, we will insert created column
df.insert(5,'bcode', NewColumn)
df
```

```python
df.loc[df['Status'] == True]
```

```python
# we made a new column " Order Status " , True = done , 
df.loc[df['Status'] == True,'Order Status'] = "done"
df.loc[df['Order Date'] <='2023-01-01', "Order Status"] = "Cancelled "
df
```

```python
# create new column called cost lever , if cost < 1000 ( high ), 
# if between 1000 - 500 it will be medium 
# if lower than 500 it will low 
def level(row): 
    if row > 1000:
        return "High"
    elif  500 <= row <= 1000:
        return "Medium"
    else: 
        return "Low"
df['Cost Level'] = df['Cost'].apply(level)
df
```

```python
import numpy as np
```

```python
df["Discount Percentage"]=np.where(df['Discount'].isnull(), "No discount",df['Discount']/df['Cost']*100)
df
```

```python
# to see unique value in any column ( without repeataion)
df["City"].unique()
```

```python
# Value counts we can count every value in the column, 
df["Country"].value_counts()
```

```python
# To sum Total Price depend on country and City 
df.groupby(["Country","City"])['Total Price'].sum()
```

```python
# To sum and mean od Total Price,depend on country and City 
df.groupby(["Country","City"])['Total Price'].agg(["sum","mean"])
```

```python
# To get diff function for diff columns grouped by Country and Cities, we should use {}> Dictionaries
df.groupby(['Country','City']).agg({"Total Price":'mean','Quantity':'sum','Cost':'max'})
```

```python
# to make pivot table that coniains row, columns and value , with functions 
df.pivot_table(index="Country",columns="Order Status",values='Quantity',aggfunc=("sum","mean"))
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df2 = pd.read_csv(r"/kaggle/input/orders/Orders.csv")
df2
```

```python
# to make a chart between Category and Counts of each on
Categories = df2['Category'].value_counts()
Categories
```

```python
# we can change the color by put,'g' for green,'r' for red #'b' for blue and 
#'g--' to make it dashed 
plt.plot(Categories,'go-')
plt.title('Categories per order', fontsize=14)
plt.xlabel('Categories', fontsize=12)
plt.ylabel('Orders')
plt.xticks(rotation=45)
# to show the chart derectly
plt.show()
```

```python
Country_Orders = df2['Country'].value_counts()
```

```python
# plt grid , alpha refers to transperancy and linestyle refers to dotted
plt.bar(Country_Orders.index,Country_Orders.values)
plt.xticks(rotation = 90)
plt.grid(axis='y',color = 'red', alpha = .35, linestyle = ':')
for ind,vl in enumerate(Country_orders):
   plt.text(ind,vl,vl, ha='center',va='bottom') # >> to put values in columns
```

```python
# next we begin to use another library 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df3 = pd.read_csv(r"/kaggle/input/orders/Orders.csv")
df3
```

```python
sns.histplot(df3['Quantity'], bins= 8)
```

```python
sns.barplot( x= 'Country', y='Total Cost',data = df3, palette = 'crest')
plt.xticks(rotation = 90 )
plt.show()
```

```python
mygraph = sns.scatterplot(x='Discount', y='Total Cost', data =df3 , 
size = 'Quantity',hue = 'Category', palette = 'crest')
mygraph.legend(bbox_to_anchor = (1,1),loc = 'upper left')
```

```python
# Index summerize 
df3.groupby(['Category','Sub Category','Item'])[['Quantity', 'Total Cost','Discount']].sum()
```

```python
# Column summerize 
df3_group = df3.groupby(['Category','Sub Category','Item'], as_index = False)[['Quantity', 'Total Cost','Discount']].sum()
df3_group
```
